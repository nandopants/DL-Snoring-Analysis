## Data Preprocessing

```python
import pandas as pd
import numpy as np
```

### Calculate Mean and standard Deviation for Every MFCC and Label

```python
def calculate_mean_std_for_list(lst):
    mean = np.mean(lst)
    std = np.std(lst)
    return pd.Series([mean, std], index=['mean', 'std'])
```

### MFCC Window Size = 500

```python
df_mfcc_500 =pd.read_csv('mfcc/all_mfcc_500ms.csv')
```

```python
df_mfcc_500
```

```python
list_columns = ['mfcc_1', 'mfcc_2', 'mfcc_3', 'mfcc_4', 'mfcc_5', 'mfcc_6', 'mfcc_7', 'mfcc_8', 'mfcc_9', 'mfcc_10', 'mfcc_11', 'mfcc_12', 'mfcc_13']
```

```python
new_data_500 = {'label': df_mfcc_500['label']}  # Create a new dictionary for the new DataFrame
```

```python
#convert string to float
for col in list_columns:
    try:
        df_mfcc_500[col] = df_mfcc_500[col].apply(eval)
    except Exception as e:
        print(f"Error in column {col}: {e}")
```

```python
for col in list_columns:
    new_col_prefix = col + '_'
    new_col_suffixes = ['mean', 'std']
    new_cols_500 = df_mfcc_500[col].apply(lambda x: calculate_mean_std_for_list(x))
    new_cols_500.columns = [new_col_prefix + suffix for suffix in new_col_suffixes]
    new_data_500.update(new_cols_500)  # Update the new data dictionary
    
# Create the new DataFrame by merging the 'Name' column and mean-std columns
new_df_500 = pd.concat([new_data_500['label']] + [new_data_500[col + '_mean'] for col in list_columns] +
                   [new_data_500[col + '_std'] for col in list_columns], axis=1)
```

```python
new_df_500.head()
```

```python
#save new data to csv file 
new_df_500.to_csv('mfcc/mfcc_500_mean_std.csv', index=False)
```
